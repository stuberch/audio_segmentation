# data analysing

```python
import librosa
import librosa.display

import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

from IPython.display import Audio, IFrame, display
```

```python
#load data
assets = r"D:\z_Downloads\Datenset_Zauberfloete\data"
x_name = "WAM-79__Track2_Channel1.wav"
x_offset = 0
x_duration = 16
x_path = assets + "\\" + x_name
x_figTitle = x_name +" (" + str(x_offset) + "s - " + str(x_offset + x_duration) + "s)"

y1_name = "WAM-100__Track2_Channel2.wav"
y1_offset = 2
y1_duration = 11
y1_path = assets + "\\" + y1_name
y1_figTitle = y1_name +" (" + str(y1_offset) + "s - " + str(y1_offset + y1_duration) + "s)"

y2_name = "WAM-46__Track1_Channel1.wav"
y2_offset = 23
y2_duration = 16
y2_path = assets + "\\" + y2_name
y2_figTitle = y2_name +" (" + str(y2_offset) + "s - " + str(y2_offset + y2_duration) + "s)"

sampleRate = 22050
x_data, sr = librosa.load(path= x_path, sr= sampleRate, offset= x_offset, duration= x_duration)
y1_data, sr = librosa.load(path= y1_path, sr= sampleRate, offset= y1_offset, duration= y1_duration)
y2_data, sr = librosa.load(path= y2_path, sr= sampleRate, offset= y2_offset, duration= y2_duration)
```

```python
#feature extraction 
hopLength = 512
hz = sampleRate / hopLength # ~2.69

x_chroma = librosa.feature.chroma_cens(x_data, sr=sampleRate, hop_length= hopLength)
y1_chroma = librosa.feature.chroma_cens(y1_data, sr=sampleRate, hop_length= hopLength)
y2_chroma = librosa.feature.chroma_cens(y2_data, sr=sampleRate, hop_length= hopLength)


#visualization
rows = 3
fig, ax = plt.subplots(nrows=rows, sharex=True, sharey=True, figsize=(20, 3 * rows))

librosa.display.specshow(x_chroma, y_axis='chroma', x_axis='s', cmap='gray', ax=ax[0])
ax[0].set(title= x_figTitle)

librosa.display.specshow(y1_chroma, y_axis='chroma', x_axis='s', cmap='gray', ax=ax[1])
ax[1].set(title= y1_figTitle)

librosa.display.specshow(y2_chroma, y_axis='chroma', x_axis='s', cmap='gray', ax=ax[2])
ax[2].set(title= y2_figTitle)

fig.subplots_adjust(hspace = 0.1 * rows)
```

```python
print(x_figTitle)
y, sr = librosa.load(path= x_path, sr= sampleRate, offset = x_offset, duration = x_duration)
display(Audio(y, rate = sr))

print(y1_figTitle)
y, sr = librosa.load(path= y1_path, sr= sampleRate, offset = y1_offset, duration = y1_duration)
display(Audio(y, rate = sr))

print(y2_figTitle)
y, sr = librosa.load(path= y2_path, sr= sampleRate, offset = y2_offset, duration = y2_duration)
display(Audio(y, rate = sr))
```
